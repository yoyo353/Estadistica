```python
import pandas as pd

Animales=pd.Series(['Gato','Tortuga','Perro','Perro','Perro','Perro','Gato','Perro','Perro'])

fi=Animales.value_counts(sort=False)
fi=fi.reindex(['Gato','Perro','Tortuga'])
tabla=pd.DataFrame({
    'fi':fi,
    'hi (%)':fi/fi.sum()*100
}).round(2)

tabla
```
