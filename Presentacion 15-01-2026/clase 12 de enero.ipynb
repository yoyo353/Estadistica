```python
import pandas as pd
df=pd.read_excel('BDActividad_1_2_3L3.xlsx')
```

```python
ocupacion=df['Ocupación']
frecuencia=ocupacion.value_counts()

tabla=pd.DataFrame({
    'fi':frecuencia,
    'hi (%)':frecuencia/frecuencia.sum()*100
}).round(2)

tabla
```

```python
import matplotlib.pyplot as plt

valores=tabla['hi (%)'].values  #Separo los valores de la frecuencia relativa
etiquetas=tabla.index #Separo los nombres de las categorías

fig,ax=plt.subplots(figsize=(6,4)) #Declaro el espacio de trabajo

ax.pie(valores,labels=etiquetas,autopct='%.1f%%',colors=['lightblue','khaki','plum','lightgreen','coral',"#E58CC2"])
ax.set_title('Distribución de Profesionales de TI según su ocupación')

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

valores=tabla['fi'].values  #Separo los valores de la frecuencia relativa
etiquetas=tabla.index #Separo los nombres de las categorías

fig,ax=plt.subplots(figsize=(6,6)) #Declaro el espacio de trabajo

columnas=ax.bar(etiquetas,valores,color=['lightblue','khaki','plum','lightgreen','coral',"#E58CC2"],edgecolor='black')
ax.bar_label(columnas)

#Personalización
ax.set_title('Distribución de Profesionales de TI según su ocupación')
ax.set_xlabel('Ocupaciones')
ax.set_ylabel('Frecuencia')
ax.set_ylim(0,30000)

plt.xticks(rotation=45,ha='right')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

edad=df['Edad']
inter=int(np.round(1+3.3*np.log10(edad.size)))
intervalos=pd.cut(edad,bins=inter,include_lowest=True,precision=2)
frecuencia=intervalos.value_counts().sort_index()

etiquetas=frecuencia.index.astype(str)
valores=frecuencia.values

fig,ax=plt.subplots(figsize=(10,6)) #Declaro el espacio de trabajo

columnas=ax.bar(etiquetas,valores,color='lightblue',edgecolor='black',width=1)
ax.bar_label(columnas)

ax.set_title('Distribución de trabajadores de TI según su edad')
ax.set_xlabel('Rangos de edad')
ax.set_ylabel('Frecuencia')


plt.xticks(rotation=45,ha='right')

plt.show()
```

```python
inter=int(np.round(1+3.3*np.log10(df['Edad'].size)))
df['Intervalos']=pd.cut(df['Edad'],bins=inter,include_lowest=True,precision=2)
frecuencia=df.groupby('Intervalos',observed=False).size()

etiquetas=frecuencia.index.astype(str)
valores=frecuencia.values

fig,ax=plt.subplots(figsize=(10,6)) #Declaro el espacio de trabajo

columnas=ax.bar(etiquetas,valores,color='lightblue',edgecolor='black',width=1)
ax.bar_label(columnas)

ax.set_title('Distribución de trabajadores de TI según su edad')
ax.set_xlabel('Rangos de edad')
ax.set_ylabel('Frecuencia')


plt.xticks(rotation=45,ha='right')

plt.show()
```
