### Tabla de frecuencia para variable **cualitativa** de un DataFrame de nombre **df**

```python
#Cálculo de frecuencia absoluta. La palabra frecuencia representa la frecuencia absoluta.
frecuencia = df.groupby('Nombre de la variable').size()

n = df.shape[0]

#Crear el DF con frecuencias absolutas y relativas
tabla = pd.DataFrame({
    'fi': frecuencia,
    'hi': round((frecuencia/n) *100, 1) # redondear según contexto y valores de los datos.
})
tabla
```

### Gráfico para variable de un DataFrame de nombre **df**


```python
#Gráfico circular

#Considerar que "tabla" es el nombre de la tabla de frecuencias construida
categorias = tabla.index  #Indices de la tabla
h = tabla['Frecuencia']   #Considerar que "Frecuencia" es el tipo de frecuencia a graficar

#Gráfico
fig,ax = plt.subplots(figsize = (6,4)) #Tamaño de la figura
circular = ax.pie(h, labels = categorias, autopct= '%.1f%%')

#Personalización
ax.set_title('Título del gráfico')

plt.show()
```

```python
#Gráfico de barras

#Considerar que "tabla" es el nombre de la tabla de frecuencias construida

#Variables
categorias = tabla.index
f = tabla['Frecuencia']   #Considerar que "Frecuencia" es el tipo de frecuencia a graficar

#Gráfico
fig, ax = plt.subplots(figsize = (6,6))
barras = ax.bar(categorias, f, color = 'r', edgecolor ='black') #Los colores son de libre elección

#Personalización
ax.set_title('Título del gráfico')
ax.set_xlabel('Título eje x')
ax.set_ylabel('Título eje y')
ax.bar_label(barras)
plt.xticks(rotation=45) #Rota en 45° las categorías para evitar la superposición
plt.show()
```

```python
#Histograma

#Considerar que "tabla" es el nombre de la tabla de frecuencias construida

#Variables
categorias = tabla.index
f= tabla['Frecuencia']   #Considerar que "Frecuencia" es el tipo de frecuencia a graficar

#Gráfico
fig, ax = plt.subplots(figsize = (7,5))
histograma= ax.bar(categorias.astype(str), f, color = 'r', edgecolor = 'black', width = 1) #Los colores son de libre elección

#Personalización
ax.set_title('Título del gráfico')
ax.set_xlabel('Título eje x')
ax.set_ylabel('Título eje y')
ax.bar_label(histograma, fmt = '%.1f%%') #El argumento fmt controla la cantidad de decimales a mostrar
plt.xticks(rotation=45, ha = 'right') #Rota en 45° las categorías para evitar la superposición
plt.show()
```
