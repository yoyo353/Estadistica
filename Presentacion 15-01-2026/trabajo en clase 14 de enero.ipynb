### Resumen

```python
#Rango - Diferencia entre el valor máximo y el mínimo

rango=altura.max()-altura.min()
rango
```

```python
#Varianza - Distancia cuadratica promedio a la media
# Mide que tal lejos están los datos en promedio de la media
varianza=altura.var()
varianza
```

```python
#Desviación estándar - Dispersión media de los datos
#Mide que tan dispersos están los datos en torno al promedio
desviacion=altura.std()
display(desviacion)

#Esto crea un rango entre el promedio-desviacion y promedio+desviacion

print(f'Se espera que la mayoría de los datos estén entre {altura.mean()-altura.std():.2f} y {altura.mean()+altura.std():.2f}')
```

```python
#Coeficiente de variación: Mide la homogeneidad de los datos - similar a la dispersión pero reescalada 

CV=altura.std()/altura.mean()*100

CV
```
