```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
lista=[x for x in range(10)]
lista_x=[1 for x in range(10)]
lista_x.append(1)
lista.append(15)

lista=np.array(lista)
```

```python
plt.scatter(lista,lista_x,zorder=2)
plt.xticks(np.arange(0,16,1))
plt.axvline(4.5,linestyle='dashed',color='red')
plt.axvline(7.5,color='green',label='Diego',linestyle='dashed')
plt.axvline(6,color='orange',label='Alfredo',linestyle='dashed')
plt.axvline(60/11,color='black',label='Media',linestyle='dashed')
plt.grid(zorder=1)
plt.legend()
plt.show()
```

```python
altura=pd.Series([1.71,1.62,1.71,1.79,1.89,1.66,1.65,1.55,1.70,1.73,1.73])

promedio=altura.mean()
print(f'La altura promedio de la muestra es {promedio:.2f}')
```

```python
mediana=altura.median()
print(f'Aproximadamente la mitad de los estudiantes mide {mediana:.2f} o menos')
print(f'Aproximadamente la mitad de los estudiantes mide {mediana:.2f} o más')
```

```python
moda=altura.mode()

print(f'La muestra tiene dos modas y corresponden a {moda[0]} y {moda[1]} ')
```
